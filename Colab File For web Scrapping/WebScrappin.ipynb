```python
!pip install selenium googletrans requests

from selenium import webdriver
from selenium.webdriver.common.by import By
from googletrans import Translator
import requests
import os

def scrape_el_pais():
    """
    Scrapes articles from the Opinion section of El País and translates the titles.
    """
    translator = Translator()
    driver = webdriver.Chrome()  #
    driver.get("https://elpais.com/")

    # Navigate to the Opinion section
    driver.find_element(By.LINK_TEXT, "Opinión").click()

    articles = driver.find_elements(By.CSS_SELECTOR, ".articulo-bloque")[:5]

    for article in articles:
        try:
            title_element = article.find_element(By.CSS_SELECTOR, "h2")
            title_spanish = title_element.text
            title_english = translator.translate(title_spanish, dest="en").text

            content_element = article.find_element(By.CSS_SELECTOR, ".articulo-cuerpo")
            content_spanish = content_element.text

            print(f"Spanish Title: {title_spanish}")
            print(f"English Title: {title_english}")
            print(f"Content: {content_spanish}")
            print("-" * 50)


            try:
                image_url = article.find_element(By.CSS_SELECTOR, ".articulo-imagen img").get_attribute("src")
                image_response = requests.get(image_url)
                image_path = f"images/{os.path.basename(image_url)}"
                with open(image_path, "wb") as f:
                    f.write(image_response.content)
            except:
                print("Image not found.")

        except Exception as e:
            print(f"Error processing article: {e}")

    driver.quit()

if __name__ == "__main__":
    scrape_el_pais()
```
